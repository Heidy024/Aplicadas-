 Paso 2: Función objetivo $Q$

La función que queremos **maximizar** es el **volumen del cilindro**:

$$
V = \pi r^2 h
$$



 Paso 3: Eliminar variables usando condiciones del problema

Usamos semejanza de triángulos del dibujo 2D:

$$
\frac{r}{H - h} = \frac{R}{H} \quad \text{(relación entre bases y alturas)}
$$

Despejamos $r$ en función de $h$:

$$
r = \frac{R(H - h)}{H}
$$

Sustituimos en la función objetivo:

$$
V(h) = \pi \left( \frac{R(H - h)}{H} \right)^2 h = \pi \frac{R^2(H - h)^2}{H^2} h
$$



 Paso 4: Encontrar puntos críticos

Derivamos $V(h)$ con respecto a $h$:

$$
V(h) = \pi \frac{R^2}{H^2}(H - h)^2 h
$$

Aplicamos regla del producto:

$$
V'(h) = \pi \frac{R^2}{H^2} \left[ -2h(H - h) + (H - h)^2 \right]
$$

Factorizamos:

$$
V'(h) = \pi \frac{R^2}{H^2} (H - h)(-2h + H - h)
$$

$$
V'(h) = \pi \frac{R^2}{H^2} (H - h)(H - 3h)
$$

Igualamos a cero:

$$
(H - h)(H - 3h) = 0
$$

Soluciones:

1. $h = H$  
2. $h = \frac{H}{3}$  



 Paso 5: Analizar los puntos críticos

* Si $h = H \Rightarrow r = 0 \Rightarrow V = 0$: no válido (no hay volumen)  
* Si $h = \frac{H}{3} \Rightarrow r = \frac{2R}{3} \Rightarrow V > 0$: válido y máximo  

Verificamos con la segunda derivada o interpretación física: el volumen es cero en los extremos, así que el máximo debe estar entre ellos.



 Resultado final

* Altura del cilindro: $\boxed{h = \dfrac{H}{3}}$  
* Radio del cilindro: $\boxed{r = \dfrac{2R}{3}}$  
* Volumen máximo:

$$
\boxed{V = \frac{4\pi R^2 H}{27}}

---

### Paso 1: Dibujo y asignación de variables

- Velocidad del pez respecto al agua: \( v \)
- Velocidad de la corriente: \( v_c \)
- Velocidad efectiva del pez respecto al suelo (contra corriente):

$$
v_{contracorriente} = v - v_c
$$

- Distancia recorrida: \( d \)
- Tiempo en recorrer \( d \):

$$
t = \frac{d}{v - v_c}
$$

---

### Paso 2: Función objetivo \( E(v) \)

La energía es proporcional al producto del tiempo y el cubo de la velocidad del pez:

$$E(v) = t \cdot v^3$$

Sustituimos el tiempo:

$$E(v) = k \cdot \frac{d}{v - v_c} \cdot v^3 = kd \cdot \frac{v^3}{v - v_c}$$

---

### Paso 3: Minimizar

Minimizamos la función:

$$f(v) = \frac{v^3}{v - v_c}$$

Derivamos usando la regla del cociente:

$$f'(v) = \frac{(3v^2)(v - v_c) - v^3(1)}{(v - v_c)^2}$$

Simplificamos el numerador:

$$f'(v) = \frac{3v^3 - 3v^2 v_c - v^3}{(v - v_c)^2} = \frac{2v^3 - 3v^2 v_c}{(v - v_c)^2}$$

Buscamos puntos críticos:

$$
2v^3 - 3v^2 v_c = 0 \\
v^2(2v - 3v_c) = 0
$$

Soluciones:

- $$ v = 0 $$ (no válida)
- $$ v = \frac{3}{2}v_c $$

---

### Conclusión:

La velocidad v que minimiza la energía gastada por el pez al nadar contra la corriente es:

$$
v = \frac{3}{2}v_c
$$



# Optimización - Varilla en un pasillo

$\textbf{Problema:} \text{Un pasillo de 6 pies de ancho da vuelta en ángulo recto.}$

 $\text{¿Cuál es la longitud de la varilla más larga que puede transportarse alrededor de la esquina, sin doblarla?}$

###Paso 1: Dibujo y variables

Supongamos que el pasillo tiene una parte horizontal y una parte vertical, ambas de 6 pies de ancho. Imaginamos una varilla recta tocando las esquinas internas mientras gira.

Denotemos:
- ( x ): la distancia desde la esquina hasta donde la varilla toca la pared en el pasillo horizontal,
- ( y ): la distancia desde la esquina hasta donde la varilla toca la pared en el pasillo vertical.

La varilla forma dos triángulos rectángulos al tocar ambas paredes, con ángulos complementarios.

---

### Paso 2: Función objetivo
Cada parte de la varilla es la hipotenusa de un triángulo rectángulo por lo tanto, la longitud total de la varilla es:

$L = \sqrt{x^2 + 6^2} + \sqrt{y^2 + 6^2}$

---

### Paso 3: Relación entre ( x ) y ( y )
En el triángulo con ángulo 𝜃, la altura es 6 y la base es x, así que:

$\tan(\theta) = \frac{6}{x} \quad \Rightarrow \quad x = 6 \cot(\theta)$
 $\tan(\theta) = \frac{y}{6} \quad \Rightarrow \quad y = 6 \tan(\theta)$

Dado que la varilla gira tocando la esquina, se cumple:

$x = 6 \cdot \cot(\theta), \quad y = 6 \cdot \tan(\theta)$

Sustituimos en la función:
$x = 6 \cdot \cot(\theta), \quad y = 6 \cdot \tan(\theta)$ en **L**

$L(\theta) = \sqrt{(6 \cot(\theta))^2 + 36} + \sqrt{(6 \tan(\theta))^2 + 36}4$

$= \sqrt{36 \cot^2(\theta) + 36} + \sqrt{36 \tan^2(\theta) + 36}$

Factorizamos 36:

$= \sqrt{36 (\cot^2(\theta) + 1)} + \sqrt{36 (\tan^2(\theta) + 1)}$

$= 6 \sqrt{\cot^2(\theta) + 1} + 6 \sqrt{\tan^2(\theta) + 1}$

Usamos identidades trigonométricas:

$cot^2(\theta) + 1 = \csc^2(\theta), \quad \tan^2(\theta) + 1 = \sec^2(\theta)$

Entonces:

$L(\theta) = 6 \csc(\theta) + 6 \sec(\theta)$

---

### Paso 4: Puntos críticos
Queremos minimizar 𝐿(𝜃), así que derivamos:
$L(\theta) = 6 \csc(\theta) + 6 \sec(\theta)$

Derivamos $L(\theta)$ utilizando las derivadas de funciones trigonométricas:

$L'(\theta) = -6 \cdot \frac{\cos(\theta)}{\sin^2(\theta)} + 6 \cdot \frac{\sin(\theta)}{\cos^2(\theta)}$

Igualamos a cero para encontrar los puntos críticos:

$-6 \cdot \frac{\cos(\theta)}{\sin^2(\theta)} + 6 \cdot \frac{\sin(\theta)}{\cos^2(\theta)} = 0$

Multiplicamos ambos lados de la ecuación por $\sin^2(\theta) \cos^2(\theta)$ para eliminar denominadores:

$-6 \cos^3(\theta) + 6 \sin^3(\theta) = 0$

$6 \sin^3(\theta) = 6 \cos^3(\theta)$

$\sin^3(\theta) = \cos^3(\theta)$

$\tan(\theta) = 1 \quad \Rightarrow \quad \theta = \frac{\pi}{4}$

---

### Paso 5: Evaluación de la longitud
Con $\theta = \frac{\pi}{4}$:

$x = 6 \cot\left( \frac{\pi}{4} \right) = 6, \quad y = 6 \tan\left( \frac{\pi}{4} \right) = 6$

Sustituyendo en la función ( L ):

$L = \sqrt{6^2 + 36} + \sqrt{6^2 + 36}
= \sqrt{72} + \sqrt{72} = 2 \cdot \sqrt{72}$

$= 2 \cdot \sqrt{36 \cdot 2} = 2 \cdot 6 \sqrt{2} = 12 \sqrt{2}$

Aproximando con $\sqrt{2} \approx 1.4142$:

$L \approx 12 \cdot 1.4142 = 16.97 \text{ pies}$

---

**Conclusión:** La varilla más larga que puede pasar por el pasillo mide aproximadamente **16.97 pies**.

```python
#1.2 Newton

# Método de Newton para resolver sin(x) = x^2
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, diff, lambdify, Eq

# Definimos la función f(x) = sin(x) - x^2
x = symbols('x')
f_expr = sin(x) - x**2
f_prime_expr = diff(f_expr, x)

# Convertimos a funciones numéricas
f = lambdify(x, f_expr, 'numpy')
f_prime = lambdify(x, f_prime_expr, 'numpy')

# Graficar f(x) = sin(x) - x^2
x_vals = np.linspace(-1.5, 1.5, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8, 4))
plt.axhline(0, color='gray', linestyle='--')
plt.plot(x_vals, y_vals, label=r'$f(x) = \sin(x) - x^2$', color='blue')
plt.title("Gráfica de $f(x) = \sin(x) - x^2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

A partir de la gráfica, podemos elegir una conjetura inicial como 𝑥 = 0.5 (observando dónde cruza el eje x).

Ahora aplicamos el método de Newton:

```python
# Método de Newton
def newton_method(f, f_prime, x0, tol=1e-4, max_iter=100):
    xi = x0
    for i in range(max_iter):
        fxi = f(xi)
        fpxi = f_prime(xi)
        if fpxi == 0:
            raise ValueError("Derivada cero. Método falla.")
        xi_next = xi - fxi / fpxi
        if abs(xi_next - xi) < tol:
            return xi_next, i+1
        xi = xi_next
    raise ValueError("No convergió en el número de iteraciones permitido")

# Ejecutar con x0 = 0.5
x0 = 0.5
root, iterations = newton_method(f, f_prime, x0)
print(f"Raíz aproximada: {root:.6f} encontrada en {iterations} iteraciones.")
```

#Resolver la ecuación $\cos(t) = t^2$ con el Método de Newton-Raphson
usando una gráfica para obtener una conjetura inicial y después aplicar el método de Newton para encontrar soluciones con un error menor a 0.0001.

---

###Paso 1: Reescribimos la ecuación
Llevamos la ecuación a la forma estándar $f(t) = 0 \$:

$f(t) = \cos(t) - t^2$

Queremos encontrar una raíz de esta función, es decir, un valor de( t ) tal que: $f(t) = 0$

---

###Paso 2: Graficar para obtener una conjetura inicial

Vamos a graficar $f(t) = \cos(t) - t^2$ para ver dónde cruza el eje ( t ) es decir, dónde se anula.




```python
import numpy as np
import matplotlib.pyplot as plt

# Dominio para visualizar la función
t = np.linspace(-2, 2, 400)
f = np.cos(t) - t**2

plt.figure(figsize=(8, 5))
plt.axhline(0, color='gray', linestyle='--')
plt.plot(t, f, label=r'$f(t) = \cos(t) - t^2$')
plt.title("Gráfica de f(t) = cos(t) - t²")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid(True)
plt.legend()
plt.show()
```

###Paso 3: Aplicar el Método de Newton
La fórmula es:

$t_{n+1} = t_n - \frac{f(t_n)}{f'(t_n)}$

Calculamos la derivada:

$f(t) = \cos(t) - t^2$

$f'(t) = -\sin(t) - 2t$
Del gráfico del paso anterior, vimos que la raíz está cerca de 0. Entonces, tomamos un valor cercano como punto de partida, por ejemplo 0.8

Cada iteración se calcula con:

$t_{n+1} = t_n - \frac{\cos(t_n) - t_n^2}{-\sin(t_n) - 2t_n}$

Ahora implementamos el método en Python:



```python
def f(t):
    return np.cos(t) - t**2

def df(t):
    return -np.sin(t) - 2*t

# Método de Newton-Raphson
def newton_method(f, df, x0, tol=1e-4, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x) / df(x)
        if abs(x_new - x) < tol:
            return x_new, i+1
        x = x_new
    return x, max_iter

# Aprox inicial desde la gráfica
t0 = 0.8
root, iterations = newton_method(f, df, t0)

print(f"Raíz aproximada: {root:.5f}, en {iterations} iteraciones")
```

## 1. $$f(x)=\dfrac{x}{x+1}$$

### Análisis sin cálculo

**Dominio y Rango:**

- Dominio: $$x+1 \ne 0 \Rightarrow x \ne -1 \Rightarrow D = \mathbb{R} \setminus \{-1\}$$

- Rango:

  $$
  y = \frac{x}{x+1} \Rightarrow y(x+1) = x \Rightarrow yx + y = x \Rightarrow x = \frac{y}{1 - y}
  $$

  Como \( x \) no está definido cuando \( y = 1 \), el rango es:

  $$
  R = \mathbb{R} \setminus \{1\}
  $$

**Simetría:**

$$
f(-x) = \frac{-x}{-x+1} \ne f(x), \quad f(-x) \ne -f(x) \Rightarrow \text{No es par ni impar}
$$

**Intersecciones:**

- Con el eje \(x\): $$f(x)=0 \Rightarrow x=0$$
- Con el eje \(y\): $$f(0)=0$$

### Análisis con cálculo

**Primera derivada:**

$$
f'(x) = \frac{(x+1)(1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2} > 0
$$

La función es creciente en todo su dominio.

**Extremos:**

No hay puntos críticos. No existen máximos ni mínimos locales.

**Segunda derivada:**

$$
f''(x) = \frac{d}{dx} \left( \frac{1}{(x+1)^2} \right) = -\frac{2}{(x+1)^3}
$$

- Cóncava hacia arriba cuando \(x < -1\)
- Cóncava hacia abajo cuando \(x > -1\)

**Asíntotas:**

- Vertical: $$x = -1$$
- Horizontal:

  $$
  \lim_{x \to \pm\infty} \frac{x}{x+1} = 1 \Rightarrow y = 1
  $$

**Asíntotas:**
